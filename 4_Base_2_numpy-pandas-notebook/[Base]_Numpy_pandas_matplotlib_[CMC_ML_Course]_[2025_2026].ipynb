```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

df_clean['publishedAt'] = pd.to_datetime(df_clean['publishedAt'], errors='coerce')
unique_videos = df_clean.drop_duplicates(subset=['video_id'])
unique_videos = unique_videos[unique_videos['publishedAt'].notna()]
count_by_date = unique_videos['publishedAt'].dt.date.value_counts().sort_index()

plt.figure(figsize=(10,5))
plt.plot(count_by_date.index, count_by_date.values)
plt.xlabel(' ')
plt.ylabel('  ( video_id)')
plt.title('   ')

plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator(maxticks=10))
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

### <font color='DarkOrange'>** 3 [1 ]**</font>

  `(scatter plot)`  -     -      (country='FR').          (comments_disabled=True).

```python
import matplotlib.pyplot as plt
import pandas as pd

df_fr = df_clean[
    (df_clean['country'] == 'FR') &
    (df_clean['comments_disabled'] == False)
].copy()

df_fr['view_count'] = pd.to_numeric(df_fr['view_count'], errors='coerce')
df_fr['comment_count'] = pd.to_numeric(df_fr['comment_count'], errors='coerce')
df_fr = df_fr.dropna(subset=['view_count','comment_count'])

#   :  95- 
vx = df_fr['view_count'].quantile(0.85)
vy = df_fr['comment_count'].quantile(0.85)
dff = df_fr[(df_fr['view_count'] <= vx) & (df_fr['comment_count'] <= vy)]

plt.figure(figsize=(8,5))
plt.scatter(dff['view_count'], dff['comment_count'], s=12, alpha=0.4)
plt.xlabel('')
plt.ylabel('')
plt.title(':  vs  (  85 )')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```
