```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img = cv2.imread('photo1.jpg')
# print(img)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# print(img_rgb)
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("")
plt.imshow(img_rgb)
plt.axis('off')
```

```python
imgText = img.copy()

cv2.putText(imgText,
            text="Hello, OpenCV",
            org=(50, 50),
            fontFace=cv2.FONT_HERSHEY_SIMPLEX,
            fontScale=1,
            color=(255, 0, 0),
            thickness=2)
plt.imshow(imgText)
```

```python
img2 = cv2.imread('photo2.jpg')
#    
# img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
# plt.imshow(img2, cmap='gray')
matrix = np.ones(img2.shape, dtype=np.uint8) * 60
imgBrighter = cv2.subtract(img2, matrix)
# plt.imshow(imgBrighter)

plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.title("")
plt.imshow(img2, cmap='gray')
plt.axis('off')

plt.subplot(1,3,2)
plt.title("")
plt.imshow(imgBrighter, cmap='gray')
plt.axis('off')

mat_low_contr = img2 *1.5
plt.subplot(1,3,3)
plt.title("")
plt.imshow(np.clip(mat_low_contr, 0, 255), cmap='gray')
plt.axis('off')
plt.show()
```

```python
img2 = cv2.imread('photo2.jpg', 0)
# 
img2 = cv2.threshold(img2, 150, 255, cv2.THRESH_BINARY)[1]
plt.imshow(img2, cmap='gray')
```

```python
#     
imgPiano = cv2.imread("photo1.jpg", 0)

#    
retval, imgPianoThresh1 = cv2.threshold(imgPiano, 50, 255, cv2.THRESH_BINARY)
retval, imgPianoThresh2 = cv2.threshold(imgPiano, 150, 255, cv2.THRESH_BINARY)

#   
imgPianoThresh3 = cv2.adaptiveThreshold(
    imgPiano, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 3
)

# 
plt.figure(figsize=(18, 5))

plt.subplot(141)
plt.imshow(imgPiano, cmap='gray')
plt.title("Orig")

plt.subplot(142)
plt.imshow(imgPianoThresh1, cmap='gray')
plt.title("Thresh1")

plt.subplot(143)
plt.imshow(imgPianoThresh2, cmap='gray')
plt.title("Thresh2")

plt.subplot(144)
plt.imshow(imgPianoThresh3, cmap='gray')
plt.title("AdaptiveThresh")

plt.show()
```
