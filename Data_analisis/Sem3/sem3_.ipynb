```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image 
import matplotlib
from IPython.display import Image
```

```python
img = cv2.imread("Apollo_11_Launch.jpg", 1)
print(img.shape)
```

```python
imgCircle = img.copy()
cv2.circle(imgCircle, (800, 500), 200, (0, 0, 255), 3);
plt.imshow(imgCircle[:,:,::-1])
```

```python
imgRect = img.copy()
cv2.rectangle(imgRect, (300, 200), (800, 600), (255, 0, 0), thickness=5)
plt.imshow(imgRect[:,:,::-1])
```

```python
imgText = img.copy()

cv2.putText(imgText, "Some test text", (200, 200), cv2.FONT_HERSHEY_SIMPLEX, 3, (0, 255, 0), 3);
plt.imshow(imgText)
```

```python
img2 = cv2.imread("New_Zealand_Coast.jpg", 1)[:,:,::-1]
plt.imshow(img2)
```

```python
matrix = np.ones(img2.shape, dtype="uint8") * 70

img2Brighter = cv2.add(img2, matrix)
img2Darker = cv2.subtract(img2, matrix)

plt.figure(figsize=[18, 5])
plt.subplot(131); plt.imshow(img2Brighter); plt.title("Brighter");
plt.subplot(132); plt.imshow(img2Darker); plt.title("Darker");
plt.subplot(133); plt.imshow(img2); plt.title("Orig");
```

```python
matLowContrast = np.ones(img2.shape) * 0.75
matHighContrast = np.ones(img2.shape) * 1.25
print(matHighContrast.dtype)

img2Low = np.uint8(cv2.multiply(np.float64(img2), matLowContrast))
img2High = np.uint8(np.clip(cv2.multiply(np.float64(img2), matHighContrast), 0, 255))

plt.figure(figsize=[18, 5])
plt.subplot(131); plt.imshow(img2Low); plt.title("Low");
plt.subplot(132); plt.imshow(img2High); plt.title("High");
plt.subplot(133); plt.imshow(img2); plt.title("Orig");
```

```python
img3 = cv2.imread("building-windows.jpg", 0)
retval, img3Thresh = cv2.threshold(img3, 150, 255, cv2.THRESH_BINARY)

plt.figure(figsize=[18, 5])
plt.subplot(121); plt.imshow(img3, cmap='gray'); plt.title("Orig");
plt.subplot(122); plt.imshow(img3Thresh, cmap = 'gray'); plt.title("Thresh");
```

```python
imgPiano = cv2.imread("Piano_Sheet_Music.png", 0)
retval, imgPianoThresh1 = cv2.threshold(imgPiano, 50, 255, cv2.THRESH_BINARY)
retval, imgPianoThresh2 = cv2.threshold(imgPiano, 150, 255, cv2.THRESH_BINARY)

imgPianoThresh3 = cv2.adaptiveThreshold(imgPiano, 255, cv2.ADAPTIVE_THRESH_MEAN_C , cv2.THRESH_BINARY, 9, 3)

plt.figure(figsize=[18, 15])
plt.subplot(221); plt.imshow(imgPiano, cmap = 'gray'); plt.title("Orig");
plt.subplot(222); plt.imshow(imgPianoThresh1, cmap = 'gray'); plt.title("Thresh1");
plt.subplot(223); plt.imshow(imgPianoThresh2, cmap = 'gray'); plt.title("Thresh2");
plt.subplot(224); plt.imshow(imgPianoThresh3, cmap = 'gray'); plt.title("Thresh3");
```
