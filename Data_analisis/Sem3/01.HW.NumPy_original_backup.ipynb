### Задача 1.4

Реализуйте функцию, которая считает проекцию одного вектора $a \in \mathbb{R}^n$ на другой вектор $b \in \mathbb{R}^n$.

````python
def project(a, b):
    a = np.asarray(a, dtype=float)
    b = np.asarray(b, dtype=float)
    denom = np.dot(b, b)
    if denom == 0:
        return np.zeros_like(a, dtype=float)
    coef = np.dot(a, b) / denom
    return coef * b

# пример
a = np.array([2, 0, 1], dtype=np.float64)
b = np.array([1, 0, 0], dtype=np.float64)
project(a, b)  # -> [2., 0., 0.]
````

````python
a = np.array([2, 0, 1], dtype=np.float64)
b = np.array([1, 0, 0], dtype=np.float64)
project(a, b)  # [2, 0, 0]
````

### Задача 1.6

Напишите функцию, создающую матрицу размера $n \times n$ с единицами на главной и побочной диагоналях.

Пример для случая 5x5:
```
array([[1, 0, 0, 0, 1],
       [0, 1, 0, 1, 0],
       [0, 0, 1, 0, 0],
       [0, 1, 0, 1, 0],
       [1, 0, 0, 0, 1]])
```

````python
def double_eye(n: int, dtype: np.dtype = np.int64):
    n = int(n)
    main = np.eye(n, dtype=dtype)
    secondary = np.fliplr(main)
    m = main + secondary
    m[m>1] = 1
    return m
````

### Задача 2.1 (1 балл)

Напишите функцию `get_nonzero_diag_product(X)`, которая подсчитывает произведение ненулевых элементов на диагонали прямоугольной матрицы. Если все элементы на диагонали нулевые, функция должна вернуть `None`.

````python
def get_nonzero_diag_product(a):
    a = np.asarray(a)
    d = np.diag(a)
    nz = d[d != 0]
    if nz.size == 0:
        return None
    return nz.prod()
````

### Задача 2.2 (1 балл)

Напишите функцию `replace_nan_to_means(X)`, принимающую матрицу $X$. Функция должна вернуть копию матрицы $X$, в которой все значения `nan` в каждом столбце заменены на среднее арифметическое остальных элементов столбца. В случае столбца из одних `nan` необходимо заменить все элементы столбца на нули. Исходная матрица `X` должна остаться неизменной.

````python
def replace_nan_to_means(X):
    X = np.asarray(X, dtype=float)
    Y = X.copy()
    col_means = np.nanmean(Y, axis=0)
    col_means = np.where(np.isnan(col_means), 0, col_means)  # столбцы полностью из NaN → 0
    i, j = np.where(np.isnan(Y))
    Y[i, j] = col_means[j]
    return Y
````

````python
assert np.isclose(
    replace_nan_to_means(np.array([[0, 1, 2], [np.nan, 1, np.nan], [5, 6, 7]])),
    np.array([[0, 1, 2], [2.5, 1, 4.5], [5, 6, 7]])).all()
assert np.isclose(
    replace_nan_to_means(
        np.array(
            [[ 6.,  1.,  5.],
             [ 6.,  8.,  7.],
             [ 5.,  8.,  7.],
             [np.nan,  2.,  2.],
             [np.nan,  3.,  1.]])
    ),
    np.array(
        [[6.     , 1.     , 5.     ],
        [6.     , 8.     , 7.     ],
        [5.     , 8.     , 7.     ],
        [5.66667, 2.     , 2.     ],
        [5.66667, 3.     , 1.     ]]
    )
).all()
````

### Задача 2.3 (1 балл)

Напишите `get_max_after_zero(x)`, возвращающую максимальный элемент в векторе $x$ среди элементов, перед которыми стоит нулевой. Если подходящих элементов нет, функция должна возвращать `None`.

````python
def get_max_after_zero(x):
    x = np.asarray(x)
    if x.size < 2:
        return None
    mask = x[:-1] == 0
    cand = x[1:][mask]
    if cand.size == 0:
        return None
    return cand.max()
````

````python
assert get_max_after_zero(np.array([1, 2, 9, 8, 0, 5])) == 5
assert get_max_after_zero(np.array([2, 0, 2, 5, 0, 1, 0, 0, 4, 7])) == 4
assert get_max_after_zero(np.array([ -7,  -1,   5,  -4,  -6,   0,  -3,  -5,   0, -10])) == -3
assert get_max_after_zero(np.array([1, 2, 3, 4, 0])) is None
````

### Задача 2.4 (1 балл)

Напишите функцию `encode_rle(x)`, реализующую кодирование длин серий (Run-length encoding). По входному вектору x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй — сколько раз их нужно повторить.

````python
def encode_rle(x):
    x = np.asarray(x)
    if x.size == 0:
        return (x, np.array([], dtype=int))
    changes = np.diff(x) != 0
    idx = np.flatnonzero(changes) + 1
    starts = np.r_[0, idx]
    ends = np.r_[idx, x.size]
    values = x[starts]
    counts = ends - starts
    return values, counts
````

````python
assert encode_rle(np.array([0])) == (np.array([0]), np.array([1]))
values, count = encode_rle(np.array([5, 5, 5, 2, 2, 2, 4, 4]))
assert (values == np.array([5,2,4])).all()
assert (count == np.array([3,3,2])).all()
values, count = encode_rle(np.array([1, 1, 1, 0, 0, 1, 1, 0, 1]))
assert (values == np.array([1, 0, 1, 0, 1])).all()
assert (count == np.array([3, 2, 2, 1, 1])).all()
````

### Задача 2.5 (1 балл)

Даны матрицы $A$ размера $(n \times d)$ и $B$ размера $(m \times d)$. Напишите функцию `find_equal_lines(A, B)`, которая найдет в A все строки, содержащиеся в B.

````python
def find_equal_lines(A, B):
    A = np.asarray(A)
    B = np.asarray(B)
    if A.size == 0 or B.size == 0:
        return np.array([], dtype=int)
    eq = (A[:, None, :] == B[None, :, :]).all(axis=2)
    mask = eq.any(axis=1)
    return np.where(mask)[0]
````

````python
A = np.array([
    [1, 2, 3],
    [1, 2, 1],
    [4, 5, 6],
    [1, 2, 1],
])
B = np.array([
    [4, 5, 6],
    [4, 5, 6],
    [1, 2, 3],
])
indices = find_equal_lines(A, B)

assert (indices == [0, 2]).all()

A = np.array([
    [0, 1],
    [0, 3],
    [0, 2],
    [1, 1],
    [0, 1],
])
B = np.array([
    [0, 1],
    [1, 1],
])
indices = find_equal_lines(A, B)

assert (indices == [0, 3, 4]).all()
````

### Задача 2.6 (1 балл)

Дана матрица $B$ размера $(n \times n)$. Напишите функцию `find_positive_columns(B)`, которая находит индексы столбцов, в которых число положительных элементов больше числа отрицательных.

````python
def find_positive_columns(B):
    B = np.asarray(B)
    pos = (B > 0).sum(axis=0)
    neg = (B < 0).sum(axis=0)
    return np.where(pos > neg)[0]
````

### Задача 2.7 (1.5 балла)

Охотники за сокровищами отправились за очередным кладом в необычное место, которое называется "Поле чудес". "Поле чудес" имеет размер $H$ на $W$ метров. На каждом квадратном метре "Поля чудес" закопан сундук с определённым количеством камней. Скоро "Поле чудес" должен посетить лепрекон. Под влиянием лепрекона камни в сундуках превращаются в золотые монеты. Лепрекон появляется в определённой клетке поля и действует на прямоугольную область размером $h$ на $w$, верхний левый край этой области - точка, где стоит лепрекон. Охотники за сокровищами не знают, когда точно появятся лепрекон, но хотят получить как можно больше золота. Известно вероятностное распределение на появление лепрекона в каждый момент времени в каждой клетке поля. Необходимо для каждой клетки "Поля чудес" посчитать математическое ожидание награды, которое будет получено от этой клетки. </br>
Необходимо написать функцию `calc_expectations(h, w, X, Q)`, где:
* $h$, $w$ - размеры области влияния лепрекона;
* $X$ - матрица целых чисел размера $H$ на $W$; $X[i, j]$ - количество камней в $i$, $j$ клетке;
* $Q$ - матрица вещественных чисел размера $H$ на $W$, задающая вероятностное распределение; $Q[i, j]$ - вероятность появления лепрекона в точке $i$, $j$.
Функция возвращает матрицу $E$ размера $H$ на $W$; $E[i, j]$ – математическое ожидание награды в $i$, $j$ клетке.

````python
def calc_expectations(h, w, X, Q):
    X = np.asarray(X, dtype=float)
    Q = np.asarray(Q, dtype=float)
    H, W = X.shape

    S = Q.cumsum(axis=0).cumsum(axis=1)

    def rect_sum(r1, c1, r2, c2):
        """Сумма элементов Q по прямоугольнику [r1:r2) × [c1:c2),
        с обрезкой по границам матрицы."""
        r1 = max(r1, 0); c1 = max(c1, 0)
        r2 = min(r2, H); c2 = min(c2, W)
        if r1 >= r2 or c1 >= c2:
            return 0.0
        total = S[r2-1, c2-1]
        if r1 > 0: total -= S[r1-1, c2-1]
        if c1 > 0: total -= S[r2-1, c1-1]
        if r1 > 0 and c1 > 0: total += S[r1-1, c1-1]
        return total

    # P[r,c] = вероятность того, что клетка (r,c) покрыта прямоугольником h×w
    P = np.empty_like(X, dtype=float)
    for r in range(H):
        for c in range(W):
            P[r, c] = rect_sum(r - h + 1, c - w + 1, r + 1, c + 1)

    return X * P
````

Выберите одну любую задачу из части 2. Реализуйте ее, не используя NumPy (можно использовать циклы `for`/`while` и т.п.). Можете сделать несколько новых реализаций. Сравните время работы каждой реализации на одинаковом наборе данных. Опишите свои наблюдения. Пример сравнения времени работы можете найти в семинаре №1.

Вывод наблюдений **нужно** сделать либо в таблице, либо на графике (например, с помощью библиотеки [matplotlib](https://matplotlib.org/), также подумайте сколько экспериментов нужно провести, чтобы сделать вывод (_спойлер_: не 1).

````python
def find_equal_lines_numpy(A, B):
    A = np.asarray(A)
    B = np.asarray(B)
    if A.size == 0 or B.size == 0:
        return np.array([], dtype=int)
    eq = (A[:, None, :] == B[None, :, :]).all(axis=2)
    return np.where(eq.any(axis=1))[0]

def find_equal_lines_loops(A, B):
    A = np.asarray(A)
    B = np.asarray(B)
    if A.size == 0 or B.size == 0:
        return np.array([], dtype=int)
    # представим строки B как хэшируемые кортежи
    set_B = {tuple(row) for row in B}
    out = []
    for i, row in enumerate(A):
        if tuple(row) in set_B:
            out.append(i)
    return np.array(out, dtype=int)
````

````python
A = np.array([
    [1, 2, 3],
    [1, 2, 1],
    [4, 5, 6],
    [1, 2, 1],
])
B = np.array([
    [4, 5, 6],
    [4, 5, 6],
    [1, 2, 3],
])

ref_np  = find_equal_lines_numpy(A, B)
ref_py  = find_equal_lines_loops(A, B)
assert (ref_np == np.array([0, 2])).all()
assert (ref_py == np.array([0, 2])).all()
assert (ref_np == ref_py).all()
````

````python
def bench_once(n=5000, m=3000, d=10, value_range=50, seed=42):
    rng = np.random.default_rng(seed)
    A = rng.integers(0, value_range, size=(n, d), dtype=np.int32)
    # часть строк A гарантированно в B, остальное случайно
    pick = rng.choice(n, size=min(n, m//2), replace=False)
    B = np.vstack([
        A[pick],
        rng.integers(0, value_range, size=(m - len(pick), d), dtype=np.int32)
    ])
    rng.shuffle(B, axis=0)

    # numpy-вариант
    t0 = perf_counter()
    idx_np = find_equal_lines_numpy(A, B)
    t1 = perf_counter()

    # python-вариант
    t2 = perf_counter()
    idx_py = find_equal_lines_loops(A, B)
    t3 = perf_counter()

    assert np.array_equal(np.sort(idx_np), np.sort(idx_py))

    return (t1 - t0), (t3 - t2), idx_np.size

cases = [
    {"n": 1000, "m": 1000, "d": 10},
    {"n": 2000, "m": 2000, "d": 10},
    {"n": 5000, "m": 3000, "d": 10},
]

print("Размеры (n, m, d) | найдено совпадений | numpy-векторизация (с) | python+set (с)")
for cfg in cases:
    tn, tp, k = bench_once(**cfg)
    print(f"{(cfg['n'], cfg['m'], cfg['d'])!s:>15} | {k:>7}             | {tn:>8.4f}            | {tp:>8.4f}")
````
