```python
import cv2
import glob
import matplotlib.pyplot as plt
import math
```

```python
imageNames = glob.glob('boat/*')
imageNames.sort()

images = []
for filename in imageNames :
    img = cv2.imread(filename, 1)[:,:,::-1]
    images.append(img)

len = len(images)
print(len)
```

```python
plt.figure(figsize=[30,10])
num_cols = 3
num_rows = int(len / num_cols)
for i in range(len):
    plt.subplot(num_rows, num_cols, i + 1)
    plt.axis('off')
    plt.imshow(images[i])
```

```python
stitcher = cv2.Stitcher_create()
retval, pano = stitcher.stitch(images)
if retval == 0:
    plt.figure(figsize=[30,10])
    plt.imshow(pano)
```
